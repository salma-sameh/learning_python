```python
#Simulate a basic ticket booking system
destination=input("pls enter your destination : ")
no_of_tickets=int(input("pls enter the number of tickets : "))
class_type=input("pls enter the class type (economy, business): ")

if class_type.lower()=='business':
    price_per_ticket=2000
else:
    price_per_ticket=1000   

total_price=no_of_tickets*price_per_ticket
print(f"The total price for {no_of_tickets} {class_type} tickets to {destination} is: ${total_price}")   

confirm=input("for booking confirmation, pls enter 'yes' or 'no' : ")
if confirm.lower()=='yes':
    print("Booking confirmed! Thank you for choosing our service.")
else:
    print("Booking cancelled. Feel free to explore other destinations with us!")
```
