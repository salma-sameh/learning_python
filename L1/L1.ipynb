```python
# string indexing and slicing
batman = 'Bruce Wayne'
print(batman[0])  # Output: 'B'
print(batman[-11])  # Output: 'B'
print(batman[10])  # Output: 'e'
print(batman[-1])  # Output: 'e'
print(batman[0:5])  # Output: 'Bruce' will include index 0 but exclude index 5
print(batman[6:])  # Output: 'Wayne' will include index 6 to the end of the string
print(batman[3:len(batman)])  # Output: 'ce Wayne' will include index 3 to the end of the string
```

```python
#Strings Reverse Slicing we write the start and end index in reverse order and add -1 as the step to reverse the string
batman = 'Bruce Wayne'
print(batman[::-1])  # Output: 'enyaW ecurB' will reverse the string
print(batman[10:0:-1])  # Output: 'enyaW e'
print(batman[10:3:-2])  # Output: 'eaWc' will include index 10 to index 4 with a step of -2
```

```python
name='salma sameh'
print(name.upper())
print(name.lower())
print(name.capitalize())
print(name.title())
print(name.swapcase()) #will swap the case of each character in the string, so uppercase letters will become lowercase and vice versa. In this case, 'salma sameh' will become 'SALMA SAMEH'.
print(name.replace('a','o'))
print(name.strip()) #The strip() method removes any leading and trailing whitespace characters from the string.
print(name.find('s')) #return the index of the first occurrence of 's' in the string, which is 0 in this case 
```
