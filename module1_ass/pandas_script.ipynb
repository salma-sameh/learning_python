```python
import pandas as pd 
import numpy as np 
```

```python
df=pd.read_csv("ex.csv")
```

```python
df.info()
```

```python
# checking the column names and the first few rows of the DataFrame
print("Initial Shape:", df.shape)
df.head()
```

```python
df.columns = (
    df.columns
      .str.strip()# Remove leading and trailing whitespace
      .str.lower()# Convert to lowercase
      .str.replace(" ", "_")# Replace spaces with underscores
)
```

```python
text_cols = df.select_dtypes(include="object").columns

for col in text_cols:
    df[col] = df[col].str.strip()# Remove leading and trailing whitespace
```

```python
df = df.drop_duplicates()
df.head(10)
```

```python
age_mapping = {
    "twenty-two": 22,
    "thirty-five": 35
}

df["age"] = df["age"].replace(age_mapping)
df["age"] = pd.to_numeric(df["age"], errors="coerce")
```

```python
df.head(10)
```

```python
#Clean Total_Purchase column
df["total_purchase"] = (
    df["total_purchase"]
      .astype(str)
      .str.replace(r"[$£,]", "", regex=True)
      .replace("free", np.nan)
)

df["total_purchase"] = pd.to_numeric(df["total_purchase"], errors="coerce")
```

```python
df["signup_date"] = pd.to_datetime(
    df["signup_date"],
    errors="coerce",
    dayfirst=True
)
```

```python
# 9️⃣ Validate Email column
email_pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"

df["email_valid"] = df["email"].str.match(email_pattern, na=False)
```

```python
#  Handle missing values
df["age"] = df["age"].fillna(df["age"].median())
df["total_purchase"] = df["total_purchase"].fillna(0)
df["country"] = df["country"].fillna("Unknown")
```
