```python
# assignment 1 answer
"""Develop a Python script that:
Takes a list of numerical values from the user
Uses functions to calculate:
Mean
Median
Mode
Implements error handling for invalid inputs
Saves results to a file stats.txt
"""
# -------- Functions --------

def mean(numbers):
    return sum(numbers) / len(numbers)


def median(numbers):
    numbers = sorted(numbers)
    n = len(numbers)

    if n % 2 == 1:
        return numbers[n // 2] # if n=5 then n//2=2 so it will finf the 3rd element in list
    else:
        return (numbers[n // 2] + numbers[n // 2 - 1]) / 2


def mode(numbers):
    freq = {}

    for num in numbers:
        freq[num] = freq.get(num, 0) + 1

    max_count = max(freq.values())

    for key, value in freq.items():
        if value == max_count:
            return key


# -------- Main Program --------

try:
    user_input = input("Enter numbers separated by commas: ")

    # Convert input to list of numbers
    numbers = [float(x) for x in user_input.split(",")]

    if len(numbers) == 0:
        raise ValueError("No numbers entered")

    # Calculate statistics
    m = mean(numbers)
    med = median(numbers)
    mo = mode(numbers)

    # Display results
    print("\nResults:")
    print("Mean:", m)
    print("Median:", med)
    print("Mode:", mo)

    # Save results to file
    with open("stats.txt", "w") as file:

        file.write("Statistics Results\n")
        file.write(f"Numbers: {numbers}\n")
        file.write(f"Mean: {m}\n")
        file.write(f"Median: {med}\n")
        file.write(f"Mode: {mo}\n")\
    
    print("\nResults saved to stats.txt")

except ValueError:
    print("Invalid input. Please enter numbers only.")

except Exception as e:
    print("Error:", e)
```
