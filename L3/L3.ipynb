```python
def absolute_value(x):
    if x>=0:
        return x
    else:
        return -x
print(absolute_value(-5))    
```

```python
# you can not modify the value of a global variable inside a function unless you use the global keyword
x="global"
def scope_check():
    x="local"
    print("x value inside the function is:",x)
scope_check()
print("x value outside the function is:",x)    
```

```python
# the global keyword allows us to modify the value of a global variable inside a function
x="global"
def scope_check():
    global x
    x="local"
    print("x value inside the function is:",x)
scope_check()
print("x value outside the function is:",x)    
```

```python
# rasise an exception
def age_check(age):
    if age<0:
        raise ValueError("age must be a positive number")
    elif age<18:
        raise ValueError("age must be at least 18")
    print("age is valid")

try:
    age_check(15)
except ValueError as e:
    print(f"error occurred: {e}")    
    
```

```python
def file_read(fname):
    with open (fname) as f:
        content_list=f.readlines()
    return content_list
print(file_read("file.txt"))
```

```python
#CreateÂ calc.py:
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        raise ValueError("Cannot divide by zero!")
    return a / b
print(add(5, 3))        # Output: 8
print(subtract(5, 3))   # Output: 2
print(multiply(5, 3))   # Output: 15
print(divide(10, 2))    # Output: 5.0
```
