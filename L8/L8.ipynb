```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
plt.plot(x, y)
plt.title('Line Plot Example')  
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# scatter plot show the relationship between two variables here we are using iris dataset which is a built in dataset in seaborn library
a=sns.load_dataset('iris')
sns.scatterplot(data=a, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot Example')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# histogram is used to show the distribution of a single variable here we are using the same iris dataset and we are showing the distribution of sepal length
sns.histplot(data=a, x='sepal_length', bins=10)
plt.title('Histogram Example')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
# pie chart is used to show the proportion of different categories here we are using the same iris dataset and we are showing the proportion of different species
species_counts = a['species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%')
plt.title('Pie Chart Example')
plt.show()
```

```python
x=[1, 2, 3, 4, 5]
y=[2, 3, 5, 7, 11]
y2=[1, 4, 9, 16, 25]
plt.subplot(1, 2, 1)# this means we are creating a 1 row and 2 columns of subplots and we are currently working on the first subplot
plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Line Plot Example 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.tight_layout()
plt.show()
```

```python
# seaborn lineplot is used to show the relationship between two variables and it also shows the confidence interval of the line here we are using the same iris dataset and we are showing the relationship between sepal length and sepal width
sns.lineplot(data=a, x='sepal_length', y='sepal_width', hue='species')
plt.title('Seaborn Line Plot Example')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# scatter plot with regression line is used to show the relationship between two variables and it also shows the regression line of the data here we are using the same iris dataset and we are showing the relationship between sepal length and sepal width
sns.lmplot(data=a, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot with Regression Line Example')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# barplot is used to show the relationship between a categorical variable and a continuous variable here we are using the same iris dataset and we are showing the relationship between species and sepal length 
sns.barplot(data=a, x='species', y='sepal_length')
plt.title('Bar Plot Example')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
#boxplot is used to show the distribution of a continuous variable and it also shows the outliers here we are using the same iris dataset and we are showing the distribution of sepal length for different species
sns.boxplot(data=a, x='species', y='sepal_length')
plt.title('Box Plot Example')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```

```python
# histogram with KDA is used to show the distribution of a single variable here we are using the same iris dataset and we are showing the distribution of sepal length
sns.histplot(data=a, x='sepal_length', bins=10, kde=True) 
plt.title('Histogram Example')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = [44, 45, 40, 41, 39]
labels = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5']
# define Seaborn color palette to use
colors = sns.color_palette('bright')

# plotting data on chart
plt.pie(data, labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```

```python
# joint plot is used to show the relationship between two variables and it also shows the distribution of each variable on the sides here we are using the same iris dataset and we are showing the relationship between sepal length and sepal width
sns.jointplot(data=a, x='sepal_length', y='sepal_width', hue='species')
plt.suptitle('Joint Plot Example', y=1.02)
plt.show()
```

```python
# violin plot is used to show the distribution of a continuous variable and it also shows the density of the data here we are using the same iris dataset and we are showing the distribution of sepal length for different species
sns.violinplot(data=a, x='species', y='sepal_length')
plt.title('Violin Plot Example')
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.show()
```
