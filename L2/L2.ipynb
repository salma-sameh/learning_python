List

```python
my_list=[1,2,3,4,5]
print(my_list)
```

```python
print(type(my_list))
```

```python
# add an element to the end of the list
my_list.append(6)
print(my_list)
```

```python
# add an element to the beginning of the list or at a specific index
my_list.insert(0,0) # insert(index,element)
print(my_list)
```

```python
# remove an element from the end of the list
out=my_list.pop()
print(out)
print(my_list)
```

```python
# remove an element by value (remove element you want to remove not the index)
my_list.remove(3)
print(my_list)
```

```python
list1=[1,2,3]
list2=[4,5,6]
# concatenate two lists
list3=list1+list2
print(list3)
```

```python
# extend a list with another list
list1.extend(list2)
print(list1)
```

```python
my_list.clear()
print(my_list)
```

```python
# copy a list
my_list=list1.copy()
print(my_list)
```

```python
# extend a list with another list
my_list.extend([9,7,8])
print(my_list)
# sort a list
my_list.sort()
print(my_list)
```

```python
print(len(my_list))
```

Set

```python
# sets contain unique elements and are unordered
set1={1,2,3,4,5}
print(set1)
```

```python
print(len(set1))
print(type(set1))
```

```python
set2=set([1,2,3,6,5])
print(set2)
print(type(set2))
print(len(set2))
```

```python
set1.add(6)
print(set1)

set1.remove(2)
print(set1)

set1.discard(10) # discard does not raise an error if the element is not found
print(set1)
```

```python
set1.remove(10) # remove raises an error if the element is not found
```

```python
set1.update(set2)
print(set1) # union of two sets taking care of duplicates
```

```python
# union of two sets returns a new set but update modifies the original set
set3=set1.union(set2)
print(set3)
```

```python
# intersection of two sets
set_intersection=set1.intersection(set2)
print(set_intersection)

set4=set1.difference(set2)
print(set4)
```

```python
set1.intersection_update(set2) # get intersection of two sets and update the original set
print(set1)
```

Dictionary

```python
# dictionaries can have keys of any immutable type and values of any type
dic1={'name':'ahmed','age':'25','gender':'male',1:[1,2,3]}
print(dic1)
```

```python
print(type(dic1))
print(len(dic1))
```

```python
dic2=dict(name='sarah',age='30',gender='female')
print(dic2)
```

```python
print(dic2['name'])
print(dic2.get('age')) # get value of a key using get method
print(dic1.get('name'))
```

```python
dic3=dic1.copy()
print(dic3)
```

```python
del dic3 # delete the dictionary
print(dic3) # this will raise an error because dic3 has been deleted
```

```python
dic_copy=dic1.copy()
dic_copy.clear() # clear the dictionary but keep the reference
print(dic_copy)
```

```python
# use fromkeys to create a dic with other dictionary keys  and values = 0
new_dic=dict.fromkeys(dic1.keys(),0)
print(new_dic)
```

```python
## use fromkeys to create a dic with other dictionary keys and values = unknown
new_dic2=dict.fromkeys(dic1.keys(),'unknown')
print(new_dic2)
```

```python
check='name' in dic1
print(check)
```

```python
print(dic1.items())
print(dic1.keys())
print(dic1.values())
```

```python
dic1.update([('name','alice'),('age','26')])# update the value of existing keys in the dictionary
print(dic1)
```

```python
dic1.update([('size','large')]) # add a new key-value pair to the dictionary using update method
print(dic1)
```
