```python
# ASSIGNMENT 1 ANSWER
"""Assignment Inventory Management System
Create a system to track products
Each product has: ID, name, quantity, price
Functions:
Add new product
Update quantity
Remove product
Display all products
Use dictionaries to store data
"""
inventory={
    1:{'name':'laptop','quantity':10,'price':1000},
    2:{'name':'phone','quantity':20,'price':500},
    3:{'name':'tablet','quantity':15,'price':300},
    4:{'name':'monitor','quantity':5,'price':200}
}
# add a new product 
inventory.update({5:{'name':'keyboard','quantity':25,'price':50}})
# update quantity of an existing product
inventory[2]['quantity']=30
inventory.update({2:{'name':'phone','quantity':30,'price':500}})
# remove a product 1
del inventory[1] # inventory.pop(1) # another way to remove a product
# display all products
for product_id, details in inventory.items():
    print(f"Product ID: {product_id}, Name: {details['name']}, Quantity: {details['quantity']}, Price: ${details['price']}")

```

```python
# ASSIGNMENT 2 ANSWER
"""Assignment 2: Implement a student gradebook with ranking and statistics.
"""
# Student Gradebook System
#  Student Dictionary 
students = {
    1: {'name': 'Salma', 'grade': 85},
    2: {'name': 'Ali', 'grade': 92},
    3: {'name': 'Mona', 'grade': 75},
    4: {'name': 'Omar', 'grade': 88}
}

#  Display Students ............
print("Student List ")
for student_id, details in students.items():
    print(f"ID: {student_id}, Name: {details['name']}, Grade: {details['grade']}")

#  Ranking Students .............
print("\nRanking ")

# *********************Convert dictionary to list of tuples*********************
student_list = list(students.items())

# Sort based on grade
student_list.sort(key=lambda x: x[1]['grade'], reverse=True)

rank = 1
for student in student_list:
    print(f"Rank {rank}: {student[1]['name']} - {student[1]['grade']}")
    rank += 1

# Statistics .................
# Create list of grades
grades_list = [details['grade'] for details in students.values()]

average_grade = sum(grades_list) / len(grades_list)
highest_grade = max(grades_list)
lowest_grade = min(grades_list)

print("\nStatistics")
print("Average Grade:", average_grade)
print("Highest Grade:", highest_grade)
print("Lowest Grade:", lowest_grade)
```
