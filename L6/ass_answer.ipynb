```python
# join to the df1 and df2 on Order ID
order_df= df1.merge(df2, on='Order ID', how='left') # how=left mean that all rows from df1 are kept, and matching rows from df2 are added where available
order_df.shape
```

```python
order_df.dropna(inplace=True)
```

```python
#Total revenue per customer
revenue_per_customer = order_df.groupby("CustomerName")["Amount"].sum().reset_index()#It converts the index back into a normal column

revenue_per_customer = revenue_per_customer.sort_values(by="Amount", ascending=False)

print(revenue_per_customer)
```

```python
#Most popular products by category
popular_products = (
    order_df.groupby(["Category", "Sub-Category"])["Quantity"]
    .sum()
    .reset_index()
)

popular_products = popular_products.sort_values(
    ["Category", "Quantity"], 
    ascending=[True, False]
)

print(popular_products)
```

```python
top_products = popular_products.loc[
    popular_products.groupby("Category")["Quantity"].idxmax()#This returns the index of the row that contains the highest quantity inside each category.
]

print(top_products)
```

```python
#Customer retention rate
orders_per_customer = order_df.groupby("CustomerName")["Order ID"].nunique()#calculates how many unique orders each customer made

repeat_customers = orders_per_customer[orders_per_customer > 1].count()#Counts customers who made more than one order
total_customers = orders_per_customer.count()#Counts total number of customers.

retention_rate = (repeat_customers / total_customers) * 100

print(f"Customer Retention Rate: {retention_rate:.2f}%")
```

```python
#Supplier performance metrics
supplier_performance = order_df.groupby("Sub-Category").agg(
    Total_Sales=("Amount", "sum"),
    Total_Profit=("Profit", "sum"),
    Total_Quantity=("Quantity", "sum")
).reset_index()

supplier_performance["Profit_Margin"] = (
    supplier_performance["Total_Profit"] /
    supplier_performance["Total_Sales"]
)

print(supplier_performance)
```
