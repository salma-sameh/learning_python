```python
import pandas as pd
df = pd.DataFrame({
'region' : ['EMEA', 'EMEA','AMER','APAC','EMEA'],
'amount': [120,90,75,60,180]

})
df['region']=df['region'].astype('category')
kpi = df.groupby('region', observed=True) ['amount']. sum().reset_index()# group by region and sum amounts, only including observed categories

print(kpi)
```

```python
agg = (df.groupby('region' ) ['amount' ]
.agg(total='sum', avg='mean', orders='count')
.reset_index())

print(agg)
```

```python
thr = 200
big = (df.groupby('region')['amount']
. sum()
.loc[lambda s: s >= thr]
.reset_index(name='total'))

print(big)
```

```python
sales = pd.DataFrame({
'month' :['2024-01', '2024-01', '2024-02', '2024-02'],
'region' : ['EMEA','AMER','EMEA','AMER'],
'amount' : [210,180, 340, 120]
})

pt = pd.pivot_table(sales, index='month', columns='region', values='amount',
aggfunc='sum', fill_value=0)
print(pt)
```

```python
pt['Total'] = pt.sum(axis=1)# this adds a new column 'Total' to the pivot table which contains the sum of values across all regions for each month
print(pt,'\n')
share = (pt.div(pt['Total'], axis=0) * 100).round(1)# this calculates the percentage share of each region's amount relative to the total for each month and rounds it to one decimal place
print(share)
```

```python
events = pd.DataFrame({'ts':pd.to_datetime(['2024-01-05','2024-01-11']),'amount':[1000, 2000]}).sort_values('ts')
fx = pd.DataFrame({'ts':pd.to_datetime(['2024-01-01','2024-01-08','2024-01-15']),
'rate': [1.1,1.12,1.09]}).sort_values('ts')

joined = pd.merge_asof(events, fx, on='ts', direction='backward')
joined['amount_local']=(joined['amount'] * joined['rate']).round(2)
print(joined)

```
